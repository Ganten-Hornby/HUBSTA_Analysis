```python
csv = pd.read_csv('/data/work/05.cluster/FuseMap/0117/tha_deg_plot/tha_sp.csv')
csv = csv[csv['logFC'] > 0.25]

merge_list = [
    'T1_NTS_CALB1',
    'T2_PRSS12',
    'T3_BMP4_IGFBP5',
    'T4_COL4A1_vessel',
    'T5_AQP4_gial_region',
    'T6_CLSTN2',
    'T7_SST_LHX_SIX3',
    'T8_CBLN4_CBLN1'
'T9_CRH',
'T10_HBZ']
var = []
for merge_li in merge_list:
    temp = csv[csv['group'] == merge_li]
    temp.sort_values('logFC', ascending = False)
    count = 0
    for i in temp.feature.tolist():
        if i in var:
            continue
        else:
            count+=1
            var.append(i)
            if count == 5:
                break
```

```python
sc.pl.matrixplot(adata, 
                 var, 
                 'dmt_leiden_annotation_0115', 
                 standard_scale = 'var',
                 # dendrogram = True,
                 show = False,
                 cmap = cmap
                )
plt.savefig(f'/data/work/05.cluster/FuseMap/0117/tha_deg_plot/sp_matrixplot_supp.pdf', bbox_inches = 'tight', dpi = 600)
plt.close()

sc.pl.dotplot(adata, 
                 var, 
                 'dmt_leiden_annotation_0115', 
                 standard_scale = 'var',
                 # dendrogram = True,
                 show = False,
                 cmap = cmap
                )
plt.savefig(f'/data/work/05.cluster/FuseMap/0117/tha_deg_plot/sp_dotplot_supp.pdf', bbox_inches = 'tight', dpi = 600)
plt.close()
```
