```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
csv = pd.read_csv('/data/work/05.cluster/FuseMap/0117/cci/GW10.csv', index_col = 'Unnamed: 0')
scaler = MinMaxScaler()
csv['lr_co_exp_ratio_scaled'] = scaler.fit_transform(csv[['lr_co_exp_ratio']])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='sr_pair', 
                y='lr_pair', 
                size='lr_co_exp_ratio', 
                hue = 'lr_co_exp_ratio',
                data=csv,
                palette = 'viridis',
               )
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
significant_points = csv[csv['is_significant'] == True]
plt.scatter(x=significant_points['sr_pair'], 
            y=significant_points['lr_pair'], 
            facecolors='none', 
            edgecolors='red', 
            s=100)  # 调整s参数以控制圆圈大小


plt.xticks(rotation=45, ha = 'right')
plt.title('GW10')
plt.savefig('/data/work/05.cluster/FuseMap/0117/cci_fig/all/GW10.pdf', bbox_inches = 'tight')
plt.close()
```
