```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

# 读取 CSV 文件
csv = pd.read_csv(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/{i}.csv', index_col='Unnamed: 0')

# 筛选 lr_pair
csv = csv[csv['lr_pair'].isin(['SST-SSTR1', 'SST-SSTR2'])]

# 对 lr_pair 按照指定顺序排序
```

```python
for i in ['GW10', 'GW12', 'GW13_rep', 'GW13', 'GW16']:
    
    csv = pd.read_csv(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/{i}.csv', index_col = 'Unnamed: 0')
    csv = csv[csv['lr_pair'].isin([
        'SST-SSTR1',
         'SST-SSTR2',
        'COL1A1-ITGA1',
        'DLL1-NOTCH1',
        'PTN-PTPRZ1',
        'MDK-PTPRZ1',
        'NRG1-ERBB4',
        'NRG3-ERBB4',
    ])]
    
    scaler = MinMaxScaler()
    csv['lr_co_exp_ratio_scaled'] = scaler.fit_transform(csv[['lr_co_exp_ratio']])

    plt.figure(figsize=(15, 3))
    sns.scatterplot(x='sr_pair', 
                    y='lr_pair', 
                    size='lr_co_exp_ratio', 
                    hue = 'lr_co_exp_ratio',
                    data=csv,
                    palette = 'viridis',
                    # legend=False  # 先关闭legend，稍后统一处理
                   )
    handles, labels = plt.gca().get_legend_handles_labels()
    labels = [f"{float(label):.2f}" for label in labels] 
    plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left')
    significant_points = csv[csv['is_significant'] == True]
    plt.scatter(x=significant_points['sr_pair'], 
                y=significant_points['lr_pair'], 
                facecolors='none', 
                edgecolors='red', 
                s=100)  # 调整s参数以控制圆圈大小


    plt.xticks(rotation=45, ha = 'right')
    plt.title(f'{i}')
    plt.savefig(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/dotplot/all_celltype/{i}.pdf', bbox_inches = 'tight')
    plt.close()
    csv = csv[csv['sr_pair'].isin([
         'mh_sc_25-mh_sc_12',
         'mh_sc_25-mh_sc_18',
         'mh_sc_25-mh_sc_19',
         'mh_sc_25-mh_sc_22',
        'mh_sc_08-mh_sc_17',
         'mh_sc_08-mh_sc_25',
         'mh_sc_08-mh_sc_26',
         'mh_sc_08-mh_sc_27',
        'mh_sc_17-mh_sc_26',

    ])]
    scaler = MinMaxScaler()
    csv['lr_co_exp_ratio_scaled'] = scaler.fit_transform(csv[['lr_co_exp_ratio']])

    plt.figure(figsize=(5, 3))
    sns.scatterplot(x='sr_pair', 
                    y='lr_pair', 
                    size='lr_co_exp_ratio', 
                    hue = 'lr_co_exp_ratio',
                    data=csv,
                    palette = 'viridis',
                    # legend=False  # 先关闭legend，稍后统一处理
                   )
    handles, labels = plt.gca().get_legend_handles_labels()
    labels = [f"{float(label):.2f}" for label in labels] 
    plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left')
    significant_points = csv[csv['is_significant'] == True]
    plt.scatter(x=significant_points['sr_pair'], 
                y=significant_points['lr_pair'], 
                facecolors='none', 
                edgecolors='red', 
                s=100)  # 调整s参数以控制圆圈大小


    plt.xticks(rotation=45, ha = 'right')
    plt.title(f'{i}')
    plt.savefig(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/dotplot/select/{i}.pdf', bbox_inches = 'tight')
    plt.close()
    
    csv = pd.read_csv(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/{i}.csv', index_col = 'Unnamed: 0')
    csv = csv[csv['lr_pair'].isin([
        'SST-SSTR1',
         'SST-SSTR2',
    ])]
    lr_order = ['SST-SSTR1', 'SST-SSTR2']
    csv['lr_pair'] = pd.Categorical(csv['lr_pair'], categories=lr_order, ordered=True)
    scaler = MinMaxScaler()
    csv['lr_co_exp_ratio_scaled'] = scaler.fit_transform(csv[['lr_co_exp_ratio']])
    lr_mapping = {lr: idx for idx, lr in enumerate(lr_order)}
    csv['lr_pair_numeric'] = csv['lr_pair'].map(lr_mapping)
    
    sr_order = sorted(csv['sr_pair'].unique())
    sr_mapping = {sr: idx for idx, sr in enumerate(sr_order)}
    csv['sr_pair_numeric'] = csv['sr_pair'].map(sr_mapping) 
    
    plt.figure(figsize=(8, 1))
    sns.scatterplot(x='sr_pair_numeric', 
                    y='lr_pair_numeric', 
                    size='lr_co_exp_ratio', 
                    hue = 'lr_co_exp_ratio',
                    data=csv,
                    palette = 'viridis',
                   )
    handles, labels = plt.gca().get_legend_handles_labels()
    labels = [f"{float(label):.2f}" for label in labels] 
    plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left')
    significant_points = csv[csv['is_significant'] == True]
    plt.scatter(x=significant_points['sr_pair'], 
                y=significant_points['lr_pair'], 
                facecolors='none', 
                edgecolors='red', 
                s=100)  
    plt.yticks(ticks=range(len(lr_order)), labels=lr_order)
    plt.ylim(-0.5, 1.5) 
    plt.xticks(ticks=range(len(sr_order)), labels=sr_order, rotation=45, ha = 'right') 
    x_min = -1
    x_max = len(sr_order)  
    plt.xlim(x_min, x_max)
    plt.title(f'{i}')
    plt.savefig(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/dotplot_sst/all_celltype/{i}.pdf', bbox_inches = 'tight')
    plt.close()
    csv = csv[csv['sr_pair'].isin([
         'mh_sc_25-mh_sc_12',
         'mh_sc_25-mh_sc_18',
         'mh_sc_25-mh_sc_19',
         'mh_sc_25-mh_sc_22',
        'mh_sc_08-mh_sc_17',
         'mh_sc_08-mh_sc_25',
     'mh_sc_08-mh_sc_26',
     'mh_sc_08-mh_sc_27',
        'mh_sc_17-mh_sc_26',

    ])]
    scaler = MinMaxScaler()
    csv['lr_co_exp_ratio_scaled'] = scaler.fit_transform(csv[['lr_co_exp_ratio']])
    sr_order = sorted(csv['sr_pair'].unique())
    sr_mapping = {sr: idx for idx, sr in enumerate(sr_order)}
    csv['sr_pair_numeric'] = csv['sr_pair'].map(sr_mapping) 
    
    plt.figure(figsize=(2, 1))
    sns.scatterplot(x='sr_pair_numeric', 
                    y='lr_pair_numeric', 
                    size='lr_co_exp_ratio', 
                    hue = 'lr_co_exp_ratio',
                    data=csv,
                    palette = 'viridis',
                   )
    handles, labels = plt.gca().get_legend_handles_labels()
    labels = [f"{float(label):.2f}" for label in labels] 
    plt.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left')
    significant_points = csv[csv['is_significant'] == True]
    plt.scatter(x=significant_points['sr_pair'], 
                y=significant_points['lr_pair'], 
                facecolors='none', 
                edgecolors='red', 
                s=100)  

    plt.yticks(ticks=range(len(lr_order)), labels=lr_order)
    plt.ylim(-0.5, 1.5) 
    
    # plt.xticks(rotation=45, ha = 'right')
    plt.xticks(ticks=range(len(sr_order)), labels=sr_order, rotation=45, ha = 'right') 
    x_min = -1
    x_max = len(sr_order)  
    plt.xlim(x_min, x_max)
    plt.title(f'{i}')
    plt.savefig(f'/data/work/05.cluster/FuseMap/0313/mh_boundary/cci/dotplot_sst/select/{i}.pdf', bbox_inches = 'tight')
    plt.close()
    
```
