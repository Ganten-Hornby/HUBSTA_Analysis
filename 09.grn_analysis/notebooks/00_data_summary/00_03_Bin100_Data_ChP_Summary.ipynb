```python
import anndata as ad
import pandas as pd
import numpy as np
import os
from pathlib import Path
from tqdm.auto import tqdm # 使用 .auto 版本以在 notebook 中获得更好的进度条
from IPython.display import display # 用于在 notebook 中漂亮地显示 DataFrame
import scanpy as sc
```

```python
annotation = pd.read_csv("fetal_brain_data_ChP/df_merged_region_annotation_mapping.csv",index_col=[0])
annotation
```

```python
annotation_Di_ChP = annotation[annotation['region_final']=='Di_ChP']
annotation_Hindbrain_ChP = annotation[annotation['region_final']=='Hindbrain_ChP']
annotation_Di_ChP
```

```python
region_folders = list(set(annotation_Di_ChP['sample_name']))
adata_list = []
for target_folder in region_folders:
    path = os.path.join('fetal_brain_hongtaoyu/0318_fetralbrain_rawcount_bin100',target_folder+'.h5ad')
    adata_obs = annotation_Di_ChP[annotation_Di_ChP['sample_name']==target_folder]
    adata = sc.read_h5ad(path)
    adata.obs_names = adata.obs_names +'|'+target_folder
    adata_list.append(adata[adata_obs.index.tolist(),:])
```

```python
combined_adata = ad.concat(
    adata_list,
    label='sample_id',     # Creates a new .obs column 'sample_id'
    join='outer'           # Keeps all genes from all files (takes the union)
)
combined_adata.write_h5ad('Process_Data/bin100_3D_region/combined_adata_Di_ChP.h5ad', compression='gzip')
```
