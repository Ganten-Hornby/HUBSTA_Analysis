# Clustering

```python
auc_mtx = sc.read_h5ad("Process_Data/specific_region/auc_hip.h5ad")
auc_mtx = auc_mtx.T
auc_mtx
```

```python
sc.pp.scale(auc_mtx)
sc.tl.pca(auc_mtx)
sc.pp.neighbors(auc_mtx)
sc.tl.leiden(auc_mtx, flavor="igraph", n_iterations=2)
auc_mtx
```

```python
auc_mtx = auc_mtx.T
auc_mtx.obs['cluster_0'] = auc_mtx[:,auc_mtx.var['leiden']=='0'].to_df().mean(axis=1)
auc_mtx.obs['cluster_1'] = auc_mtx[:,auc_mtx.var['leiden']=='1'].to_df().mean(axis=1)
auc_mtx.obs['cluster_2'] = auc_mtx[:,auc_mtx.var['leiden']=='2'].to_df().mean(axis=1)
auc_mtx.obs['cluster_3'] = auc_mtx[:,auc_mtx.var['leiden']=='3'].to_df().mean(axis=1)
auc_mtx.obs['cluster_4'] = auc_mtx[:,auc_mtx.var['leiden']=='4'].to_df().mean(axis=1)
auc_mtx.obs['cluster_5'] = auc_mtx[:,auc_mtx.var['leiden']=='5'].to_df().mean(axis=1)
auc_mtx
```

```python
for _slice in set(auc_mtx.obs['slice_code']):
    sc.pl.spatial(auc_mtx[auc_mtx.obs['slice_code']==_slice],color=['cluster_0','cluster_1','cluster_2','cluster_3','cluster_4','cluster_5'],spot_size=50,vmax='p99')
```
